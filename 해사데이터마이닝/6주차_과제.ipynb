### 💡 More Study 1. 초간단 템플릿의 '결과 해석 및 액션 플랜'

코드를 실행한 뒤 나오는 결과물을 보고 다음 단계(전처리)를 어떻게 결정할지 연결하는 가이드입니다.

**1. 결측치 (Missing Values)**
* **해석:** 특정 변수에 결측치가 얼마나 집중되어 있는가?
* **액션:**
    * **5% 미만:** 해당 행(Row) 삭제 또는 평균/중앙값/최빈값으로 단순 대체
    * **10~50%:** 다른 변수들을 활용해 결측치를 예측해서 채우거나(KNN Imputer 등), 비즈니스 로직에 따라 대체
    * **50% 이상:** 해당 변수(Column) 자체를 삭제하는 것을 고려 (단, 결측된 것 자체가 의미를 가지는 경우 예외)

**2. 자료형 (Data Types)**
* **해석:** 숫자로 되어있지만 실제로는 범주형(고객 등급: 1, 2, 3)인 변수가 없는가? 반대로 문자열로 되어있지만 실제로는 숫자(가격: "1,000달러")인 변수가 없는가?
* **액션:** 숫자형인 척하는 범주형 변수는 `astype('object')`나 `astype('category')`로 변환하고, 범주형 변수(문자)는 원핫 인코딩(One-Hot)이나 라벨 인코딩을 적용할 준비를 합니다.

**3. 기초통계 (Summary Statistics)**
* **해석:** `min`과 `max` 값이 정상 범주에 있는가? `mean`(평균)과 `50%`(중앙값)의 차이가 큰가?
* **액션:**
    * 평균과 중앙값 차이가 크다면 데이터가 한쪽으로 치우친(Skewed) 상태입니다. -> **로그 변환(Log Transform)** 고려
    * 상식적으로 말이 안 되는 `min`, `max` (예: 나이가 999살, 가격이 음수) -> **이상치(Outlier) 제거 또는 대체**

**4. 타깃 분포 (Target Distribution)**
* **해석:** 분류(Classification) 문제에서 정답(Target) 클래스의 비율이 8:2 또는 9:1 이상으로 심하게 차이 나는가?
* **액션:** 데이터 불균형(Imbalance)이 심하다면 정확도(Accuracy)는 의미가 없어집니다. -> **SMOTE 같은 오버샘플링 적용, 모델 학습 시 `class_weight='balanced'` 부여, 평가지표로 F1-Score나 ROC-AUC 사용**

**5. 상관관계 (Correlation)**
* **해석:** 타깃 변수와 상관관계가 높은 변수(좋은 피처)는 무엇인가? 독립변수들끼리 상관관계가 너무 높은(0.8 이상) 쌍이 있는가?
* **액션:**
    * 독립변수 간 상관관계가 너무 높으면 다중공선성(Multicollinearity) 문제가 발생해 모델의 해석력을 떨어뜨립니다. -> **둘 중 하나를 삭제하거나 PCA(주성분 분석)로 차원 축소**
